Encapsulation In Python

```python
class address:
  def __init__(self , city , state):
    self.city = city
    self.state = state
  def location(self):
    return f"my location is {self.city} in {self.state}"

a = address("Chappra" , "Bihar")
print(a.location())
print(a.city)
print(a.state)


print('\n')
print("____________________")
print('\n')


class Address:
  def __init__(self,city,state):
    self.__city = city # private attribute
    self.state = state

  def location(self):
    print(f"my location is {self.__city} in {self.state}")

  def getcity(self):
    return self.__city

a = Address("chhapra","Bihar")
a.location()
print(a.getcity())
#print(a.__city)
print(a.state)
```
