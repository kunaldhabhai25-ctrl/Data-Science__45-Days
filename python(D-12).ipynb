```python
import pandas as pd
d = {
     "Name" : ["vishal" , "virat" , "vinod"] ,
     "salary" : [1000,2000,3000]
}
df = pd.DataFrame(d)
print(df)
print('\n')
print("-----------------------")
#add column
df["marks"] = [10,20,30]
print(df)
print('\n')
print("-----------------------")
#copy data
df["marks"] = df["salary"]
print(df)
#adding column
print('\n')
print("-----------------------")
df["Increament(10%)"] = df["salary"]+df["salary"]/10
print(df)
#applying changes in main data
print('\n')
print("-----------------------")
df.rename(columns={"increament(10%)" : "Salary Increment(10%)"} , inplace = True)
print(df)

```

#Drop

```python
import pandas as pd
d = {
     "Name" : ["vishal" , "virat" , "vinod"] ,
     "salary" : [1000,2000,3000]
}
df = pd.DataFrame(d)
#add column
df["marks"] = [10,20,30]
print(df)
print('\n')
print("-----------------------")
#drop column
print('\n')
df.drop("marks" , axis = 1)#no change in main Data file
print(df)
print('\n')
print("-----------------------")
df.drop("marks" , axis = 1 , inplace = True)#to make changes in the original data file
print(df)


```

#Adding And Deleting Rows'

```python
#iloc --> index basec fetching of data
#loc --> lable based fetching of data
import pandas as pd
d = {
     "Name" : ["vishal" , "virat" , "vinod"] ,
     "salary" : [1000,2000,3000]
}
df = pd.DataFrame(d)
df
print("--Using Loc--")
print(df.loc[2,"Name"])
print("--Using  iLoc--")
print(df.iloc[2,0])

#getting single column using iloc
print("--Single Column (iLoc)--")
print(df.iloc[1])

#getting single column using loc
print("--Single Column (Loc)--")
print(df.loc[1])

#for multiple rows
print("--Multiple Rows (iloc)--")
print(df.iloc[0:2])

print("--Multiple Rows (loc)--")
print(df.loc[0:2])

#getting sub data
print("--sub data (iloc)--")
print(df.iloc[0:2 , 0])

print("--sub data (loc)--")
print(df.loc[0:2 ,"salary"])
```
