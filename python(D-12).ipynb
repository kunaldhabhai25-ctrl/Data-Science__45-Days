#Getting Data Or Adding Columns

```python

import pandas as pd
print("_____JSON_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file2.json"
df = pd.read_json(url)
print(df)
print('\n')
df["name"]
#add single column
print('\n')
print("____________________________________________")
print('\n')
df["salary"] = df["marks"]*100
print(df)
#alternate way
print('\n')
print("____________________________________________")
print('\n')
df["salary"] = [10,20,30,40,600]
print(df)

```

```python
import pandas as pd
d = {
     "Name" : ["vishal" , "virat" , "vinod"] ,
     "salary" : [1000,2000,3000]
}
df = pd.DataFrame(d)
print(df)
print('\n')
print("-----------------------")
#add column
df["marks"] = [10,20,30]
print(df)
print('\n')
print("-----------------------")
#copy data
df["marks"] = df["salary"]
print(df)
#adding column
print('\n')
print("-----------------------")
df["Increament(10%)"] = df["salary"]+df["salary"]/10
print(df)
#applying changes in main data
print('\n')
print("-----------------------")
df.rename(columns={"increament(10%)" : "Salary Increment(10%)"} , inplace = True)
print(df)

```

#Drop

```python
import pandas as pd
d = {
     "Name" : ["vishal" , "virat" , "vinod"] ,
     "salary" : [1000,2000,3000]
}
df = pd.DataFrame(d)
#add column
df["marks"] = [10,20,30]
print(df)
print('\n')
print("-----------------------")
#drop column
print('\n')
df.drop("marks" , axis = 1)#no change in main Data file
print(df)
print('\n')
print("-----------------------")
df.drop("marks" , axis = 1 , inplace = True)#to make changes in the original data file
print(df)


```
