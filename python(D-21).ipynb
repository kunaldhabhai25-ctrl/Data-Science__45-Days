#Feature Scaling

```python
from matplotlib.lines import lineStyles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = {
    "exprience" : [300,600,900,1500] ,
    "salary" : [1000,1500,2000,3000]
}
df = pd.DataFrame(data)
print("                    -Original Data-                    ")
print(df)
print('\n')
#standard scaler
scaler = StandardScaler()
df["exprience"] = scaler.fit_transform(df[["exprience"]])
print("                    -Scaled Data-                    ")
print(df)
print('\n')
#split data
X = df['exprience']
Y = df['salary']
print("                    -Split Data-                    ")
print(X)
print(Y)
print('\n')
plt.plot(X,Y,linestyle ='--',marker = "o")
plt.xlabel("exprience")
plt.ylabel("salary")
plt.title("DATA")
plt.grid()
plt.show()
```

#Data Split Training Testing

```python
from matplotlib.lines import lineStyles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

data = {
    "exprience" : [300,600,900,1500] ,
    "salary" : [1000,1500,2000,3000]
}
df = pd.DataFrame(data)
print("                    -Original Data-                    ")
print(df)
print('\n')

#Split Data
x = df["exprience"]
y = df["salary"]

#Train And Test Split
x_Train , x_Test , y_Train , y_Test = train_test_split(x,y,test_size =0.2 , random_state=2)
print("                    -Train Data-                    ")
print(x_Train)
print(y_Train)
print('\n')
print("                    -Test Data-                    ")
print(x_Test)
print(y_Test)
print('\n')
```
