#Feature Scaling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = {
    "exprience" : [300,600,900,1500] ,
    "salary" : [1000,1500,2000,3000]
}
df = pd.DataFrame(data)
print("                    -Original Data-                    ")
print(df)
print('\n')
#standard scaler
scaler = StandardScaler()
df["exprience"] = scaler.fit_transform(df[["exprience"]])
print("                    -Scaled Data-                    ")
print(df)
print('\n')
#split data
X = df['exprience']
Y = df['salary']
print("                    -Split Data-                    ")
print(X)
print(Y)
```
