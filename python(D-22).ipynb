```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = {
    "Name" : ["Kunal","Dheeraj","Yash","Ankita"],
    "House_Age" : [20.2,2.5,3.0,4.0],
    "Avg_Income" : [1000,200,300,400],
    "Num_Rooms" : [3,4,5,1],
    "Price" : [10000,70000,15000,11000]
}
#data frame
print("                      -Original Data-                      ")
df = pd.DataFrame(data)
print(df)
print('\n')
#handle missing values
print("                      -Data After Data Cleaing-                      ")
df.drop("Name" , axis = 1 , inplace = True)
print(df)
print('\n')
#feature scaling
print("                      -Data After Feature Scaling-                      ")
scaler = StandardScaler()
df[["Avg_Income","House_Age","Num_Rooms"]] = scaler.fit_transform(df[["Avg_Income","House_Age","Num_Rooms"]])
print(df)
print('\n')
#split data
x = df[["Avg_Income","House_Age","Num_Rooms"]]
y = df["Price"]
#train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 ,random_state = 2)
print("                      -Data After Train Test Split-                      ")
print("_______________________X_Train_Test_______________________")
print(x_train)
print(x_test)
print("_______________________Y_Train_Test_______________________")
print(y_train)
print(y_test)
print('\n')
#model train
model = LinearRegression()
model.fit(x_train,y_train)
#model pridiction
avg_income = int(input("Enter Average Income : "))
house_age = float(input("Enter House Age : "))
num_rooms = int(input("Enter Number of Rooms : "))
new_data = {
    "Avg_Income" : [avg_income],
    "House_Age" : [house_age],
    "Num_Rooms" : [num_rooms]
}
new_df = pd.DataFrame(new_data)
new_df[["Avg_Income","House_Age","Num_Rooms"]] = scaler.transform(new_df[["Avg_Income","House_Age","Num_Rooms"]])
predicted_price = model.predict(new_df)
print("_______________________Predicted Price_______________________")
print("Predicted Price : ", predicted_price[0])

#model dump
# !pip install joblib
import joblib
joblib.dump(model,"House_Model.joblib")

```
