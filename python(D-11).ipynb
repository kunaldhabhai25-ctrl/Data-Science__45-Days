Question --> Apply All The Above Concepts On Given Data

```python
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
print("---------------Question---------------")
df = pd.DataFrame(data)

print(df)
print('\n')
#shape
print("_____SHAPE_____")
print(df.shape)
#rename
print('\n')
print("_____RENAME_____")
df.rename(columns={"Name":"Emp_Name"},inplace=True)
print(df)
print('\n')
#info
print("_____INFO_____")
df.info()
print('\n')
#head
print("_____HEAD_____")
print(df.head(2))
print('\n')
#tail
print("_____TAIL_____")
print(df.tail(2))
print('\n')
#describe
print("_____DESCRIBE_____")
df.describe()

```
