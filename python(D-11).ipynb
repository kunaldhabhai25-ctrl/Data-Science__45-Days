#Save And Load Data From CSV , JSON And Excel


```python
#CSV
import pandas as pd
print("_____CSV_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file1.csv"
df = pd.read_csv(url)
print(df)
print('\n')
# from google.colab import drive
# drive.mount('/content/drive')
# url = "/content/drive/MyDrive/file1.csv"
# df = pd.read_csv(url)
# print(df)


#Json
print("_____JSON_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file1.json"
df = pd.read_json(url)
print(df)
print('\n')
#head
df.head(4)
print('\n')
#tail
df.tail(3)
print('\n')



#Excel
print("_____EXCEL_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file1.xlsx"
df = pd.read_excel(url)
print(df)




```

#Head() and Tail() Function

```python
import pandas as pd
print("_____CSV_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file1.csv"
df = pd.read_csv(url)
print(df)
print('\n')
#head
df.head()
df.head(-2)

#tail
df.tail()
df.tail(2)
df.tail(-2)

# only prints the last one if we not use print

```

```python
import pandas as pd
print("_____CSV_____")
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file1.csv"
df = pd.read_csv(url)
print(df)
print('\n')
#head with print
print("---HEAD---")
print(df.head(4))
print('\n')
#tail wirh print
print("---TAIL---")
print(df.tail(4))

# both can be displayed using print
```
