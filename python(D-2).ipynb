FUNCTIONS IN PYTHON


```python
#function
#example no. --> 1
def hello():
  print("hello world!")

hello()

print ("---------------")

#example no. --> 2
def hello2(a): #parameter
  print(a)

hello2(199) #argument

print ("---------------")

#example no. --> 3
def sum(num1,num2):
  print("the sum of ", num1 , "+" , num2 , "is :", num1 + num2)

sum(10,20)

print ("---------------")

#example no. --> 4
def power(num1,num2):
  print(num1**num2)

power(2,3)
power(num1=6,num2=3)

print ("---------------")

#example no. --> 5

def student(*a):
  print(a)
  print(type(a))

student(1 , 2 , 3 , 4 , 5 , 6 , 7 )


```

```python
#Lambda function in Python [lambda argument : expression]

add = lambda x: x
print(add(10))

print ("---------------")

add = lambda x,y : x + y
print(add(10,20))
```

```python
#Lambda function with star parameter

add = lambda *a : a[0]+a[1]
print(add(10,20))
```

```python
#list compression [expression for item in iterable]


#example
print("----------EXAMPLE----------")
L = [10,20,30,40,50,60]
print([L[i] for i in range(len(L))])


print("----------QUESTION----------")

#question
L = [10,20,[30,40,50,60]]
print([L[2][i] for i in range(len(L))])

```

```python
#LIST
#initialization
L = [10,20,30,40]

print("length of list is : " , len(L))

print("---------------")

#append in list
L.append(50)
print("-----LIST AFTER APPENDING-----")
print(L)

#insert in list
L.insert(5,60)
print("-----LIST AFTER INSERTING-----")
print(L)

#remove in list
L.remove(60)
print("-----LIST AFTER REMOVING-----")
print(L)

print("-----LIST ELEMENTS-----")

#accessing individual elements of list
for i in range(len(L)):
  print("Element at index " , i , "is :" , L[i])

```

Question No. --> 1 Of List


```python
L = [10,20,30,{"name" : "kunal" , "address" : ["jaipur","kukas","hometown","friends home"]}]
print("---Without Loop---")
print(L[-1])
print("---Using Loop---")
for i in L[-1]["address"]:
  print(i, end = " ")
```

```python
#DICTIONARY IN PYTHON
dict = {"name" : "kunal" , "age" : "21"}
print(dict)

print(dict.keys())
print(dict.values())
for i in dict.keys():
  print("key = ", i)
  print("value = ", dict[i])

```
