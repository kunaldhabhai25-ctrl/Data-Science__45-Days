```python
import pandas as pd
import matplotlib.pyplot as plt
#reading json file
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-arya/main/temperature_data.json"
df=pd.read_json(url)
print("                              --Question--                              ")
print(df)
#data cleaning
df.dropna(subset=["temperature_c"], inplace=True)
df.fillna(df["humidity_pct"].mean(),inplace=True)
print("____________________________________________________________________")
print('\n')
print("                              --Data Cleaning--                              ")
print(df)
print("____________________________________________________________________")
print('\n')
print("                              --Adding New Column--                             ")
df["Fahrenheit"] = df["temperature_c"]*1.8+32
print(df)
#to reset the index
print("____________________________________________________________________")
print('\n')
print("                              --Reset Index--                             ")
df.reset_index(drop=True, inplace=True)
print(df)
print('\n')
print("____________________________________________________________________")
#plotting graphs
fig , ax = plt.subplots(1,2)
ax[0].plot(df["temperature_c"],df["humidity_pct"],linestyle = "dotted",marker="o",markeredgecolor="orange")
ax[0].set_xlabel("Temperature")
ax[0].set_ylabel("Humidity")
ax[0].set_title("Temperature vs Humidity")
ax[1].pie(df["temperature_c"],labels=df["day"],explode=(0.1,0,0,0,0,0),shadow = True)
ax[1].set_title("Temperature On Days")


plt.tight_layout()

plt.show()


```
