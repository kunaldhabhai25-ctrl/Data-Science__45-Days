```python
from matplotlib import lines
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,30,40]
y2=[20,30,25,30]
y3=[11,22,57,10]
w = 0.20
plt.bar(x-w,y1,label = "Boys",width = w)
plt.bar(x,y2,label = "Girls" , width = w)
plt.bar(x+w,y3,label = "Total",width = w)
plt.legend()
plt.xlabel("Group")
plt.ylabel("No. of Students")
plt.title("Grouped Bar Chart")
plt.show()
```

#Histograph

```python
import matplotlib.pyplot as plt
marks = [40,55,60,70,75,90,33,50]
plt.hist(marks,bins=2,color="pink",orientation="vertical")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram")
```

#Pie Chart

```python
import matplotlib.pyplot as plt
fruits = ["Apple","Banana","Orange","Mango","Strawberry"]
count = [40,30,15,78,45]
plt.plot(fruits , count)
plt.pie(count,labels=fruits,startangle=90,explode=(0.1,0,0,0,0),shadow=True)
plt.title("Pie Chart")

plt.show()
```
