```python
from matplotlib import lines
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,30,40]
y2=[20,30,25,30]
y3=[11,22,57,10]
w = 0.20
plt.bar(x-w,y1,label = "Boys",width = w)
plt.bar(x,y2,label = "Girls" , width = w)
plt.bar(x+w,y3,label = "Total",width = w)
plt.legend()
plt.xlabel("Group")
plt.ylabel("No. of Students")
plt.title("Grouped Bar Chart")
plt.show()
```

#Pie Chart

```python
import matplotlib.pyplot as plt
fruits = ["Apple","Banana","Orange","Mango","Strawberry"]
count = [40,30,15,78,45]
plt.plot(fruits , count)
plt.pie(count,labels=fruits,startangle=90,explode=(0.1,0,0,0,0),shadow=True)
plt.title("Pie Chart")

plt.show()
```

#Subplots'

```python
#subplots (1D)
import matplotlib.pyplot as plt
#data for graph-1
year = [2010,2015.2020,2025]
dairy = [100,500,620,400]
#data for graph-2
year = [1990,2000,2005,2015]
farming = [300,200,250,100]

fig , ax = plt.subplots(1,2)

ax[0].plot(year , dairy)
ax[0].set_title("Dairy Production")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Production")

ax[1].plot(year , farming)
ax[1].set_title("Farming Production")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Production")

plt.tight_layout()
plt.show()
```

```python
#subplots (2D)
import matplotlib.pyplot as plt
#data for graph-1
year = [2010,2015.2020,2025]
dairy = [100,500,620,400]
#data for graph-2
year = [1990,2000,2005,2015]
farming = [300,200,250,100]
#data for graph-3
year = [2010,2015.2020,2025]
college_students = [100,200,300,600]
#data for graph-4
year = [1990,2000,2005,2015]
transport = [50,100,250,500]


fig , ax = plt.subplots(2,2)

ax[0][0].pie(year,labels=dairy,startangle=90,explode=(0.1,0,0,0),shadow=True)
ax[0][0].set_title("Dairy Production")
ax[0][0].set_xlabel("Year")
ax[0][0].set_ylabel("Production")

ax[0][1].plot(year , farming, color ="red" , linestyle = "dotted" , marker = "o")
ax[0][1].set_title("Farming Production")
ax[0][1].set_xlabel("Year")
ax[0][1].set_ylabel("Production")

ax[1][0].plot(year , college_students , color = "green")
ax[1][0].set_title("College Students")
ax[1][0].set_xlabel("Year")
ax[1][0].set_ylabel("Students")

ax[1][1].bar(year , transport)
ax[1][1].set_title("Transport")
ax[1][1].set_xlabel("Year")
ax[1][1].set_ylabel("Transport")

plt.tight_layout()


plt.gcf().canvas.get_supported_filetypes()
plt.savefig("Subplots.png")
plt.show()
```
