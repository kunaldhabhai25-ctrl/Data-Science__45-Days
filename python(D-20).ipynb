#Data Encoding
1) Label Encoding
2) One Hot Encoding

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
data ={"color": ['red', 'green', 'yellow', 'green','red',np.nan]}
df=pd.DataFrame(data)
print("          -original data-          ")
print(df)
print('\n')

#handle null values
df.dropna(inplace=True)
print("          -data after handling null values-          ")
print(df)
print('\n')

print("          -data after label encoding-          ")

#label encoding
le=LabelEncoder()
df['colors_encoder']=le.fit_transform(df['color'])
print(df)
df.drop(["colors_encoder"],axis=1,inplace=True)
print('\n')

print("          -data after one hot encoding-          ")
#one hot encoding(by th help of sklearn)
# encoder = OneHotEncoder()
# encoder_df = encoder.fit_transform(df[["color"]])
# print(encoder_df.toarray())

#one hot encoding(by the help of pandas)
df=pd.get_dummies(df,columns=['color'])
print(df)



```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
data ={"Age":[20,25,np.nan,30,22],
       "Gender":["Male" , "Female" , "Others" , "Male" , "Female"],
       "Name": ["Dheeraj" , "Anjali" , "Ali" , "Kunal" , None]
       }
df=pd.DataFrame(data)
print("          -original data-          ")
print(df)
print('\n')
#handelling missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
print("          -data after handling missing values-          ")
print(df)
print('\n')
#removing rows with missing values
df.dropna(subset = ["Name"] , inplace = True)
print("          -data after removing rows with missing values-          ")
print(df)
print('\n')
#Label Encoding
le = LabelEncoder()
df["Gender_Encoder"]=le.fit_transform(df["Gender"])
print("          -data after label encoding-          ")
print(df)
print('\n')
df.drop(["Gender_Encoder"],axis=1,inplace=True)
#One Hot Encoding(Using Sklearn)
print("          -data after one hot encoding(Using Sklearn)-          ")
encoder = OneHotEncoder()
encoder_df = encoder.fit_transform(df[["Gender"]])
print(encoder_df.toarray())
print('\n')
#One Hot Encoding(Using Pandas)
df=pd.get_dummies(df,columns=["Gender"])
print("          -data after one hot encoding(Using Pandas)-          ")
print(df)
```
