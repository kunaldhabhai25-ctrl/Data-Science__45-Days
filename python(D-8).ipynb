Question --> Insert 100 At place of 4

```python
#List
print('\n')
print("-----List Case-----")
print('\n')
L = [[1,2,3],[4,5,6]]
L[1].insert(0,100)
print(L)



import numpy as np
#Array
print('\n')
print("-----Array Case-----")
print('\n')
arr = np.array([[1,2,3],[4,5,6]])
arr = np.insert(arr, 0, 100, axis=1)

print(arr)

# in list inserting 100 only at specific place is possible , but in array the size must remain same for both rows and columns. so,that's why insertion of only 100
# is not possible in array. so, to make it work we need to add elements in both row and column
```

#Zero's In Python





```python
#zero's array in 1D
import numpy as np
print("Zero's Array In 1D")
arr = np.zeros(5)
print(arr)
print('\n')
#zero's array in 2D
print("Zero's Array In 2D")
arr = np.zeros((2,3))
print(arr)

#zero's array in 3D
print('\n')
print("Zero's Array In 3D")
arr = np.zeros(((2,3,4)))
print(arr)


```

#One's In Python

```python
#one's array in 1D
import numpy as np
print("--One's Array In 1D--")
arr = np.ones(5)
print(arr)
print('\n')
#one's array in 2D
print("--One's Array In 2D--")
arr = np.ones((2,3))
print(arr)

#one's array in 3D
print('\n')
print("--One's Array In 3D--")

arr = np.ones(((2,3,4)))
print(arr)
```

#Full In Python

```python
#full for id
import numpy as np
arr = np.full((2),10) #1-D
print("--1-D Array Using Full--")
print(arr)
print('\n')

arr = np.full((2,3),10) #2-D
print("--2-D Array Using Full--")
print(arr)
print('\n')

arr = np.full((2,3,4),10) #3-D
print("--3-D Array Using Full--")
print(arr)
```
