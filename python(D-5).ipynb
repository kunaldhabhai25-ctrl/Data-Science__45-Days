```python
from os import name
#example n0. --> 1
print("----------Example no. --> 1----------")
class kunal:
  def __init__(self,name):
    self.name = name
  def show(self):
    print(self.name)

r = kunal("hello world!!!")
r.show()


print('\n')
print("____________________")
print('\n')
#example no. --> 2
print("----------Example no. --> 2----------")
class kunal:
  def __init__(self):
    print("calling constructor")
  def show(self):
    print("show the name")

k = kunal()
k.show()

print('\n')
print("____________________")
print('\n')

#example no. --> 3
print("----------Example no. --> 3----------")
class kunal:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def getName(self):
    print("My Name Is : " , self.name)

  def getAge(self):
    print("My Age Is : " , self.age)


k = kunal("kunal",22)
k.getName()
k.getAge()

print('\n')
print("____________________")
print('\n')

#example n0. --> 4
print("----------Example no. --> 4----------")
class student:
  def __init__(self,*args):
    print(type(args))
    self.name = args
  def getStd(self):
    print("The Student Name Is : ",self.name)


s = student("kunal" , "999000xxxx" , 20 ," kxxxxx@gmail.com" )
s.getStd()


print('\n')
print("____________________")
print('\n')

#example no. --> 5
print("----------Example no. --> 5----------")
class classroom:
  def __init__(self,args):
    print(type(args))
    print(args)
    self.name = args

  def getData(self):
    return self.name

c = classroom({"name":"kunaal" , "Age":"21"})
print(c.getData()["name"])
print(c.getData()["Age"])


print('\n')
print("____________________")
print('\n')
#example no. --> 6
print("----------Example no. --> 6----------")
class student:
  def __init__(self,*args):
    self.data = args

  def users(self):
    for i in self.data[0]:
      print(i)

  def details(self):
    for i in self.data[1]:
      print(i , "=" , self.data[1][i])

s = student(["kunaal" , "dheeraj" , "harsh" , "praveen"] , {"address" : "kukas" , "college": "Arya" , "loc" : "jaipur"})
s.users()
s.details()

```
