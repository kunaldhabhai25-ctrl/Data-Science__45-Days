# Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans

data = {
    "Age" : [22,25,24,45,50,48,20,47],
    "Income" : [22000,25000,24000,45000,50000,48000,20000,47000]
}
df = pd.DataFrame(data)
print("                  DATA                  ")
print(df)

#input | feature
X = df[["Age" , "Income"]]

#model
model = KMeans(n_clusters=2,random_state=42)
model.fit(X)

#clusters
clusters = model.labels_
print("                      -Clusters'-                      ")
print(clusters)
print("                    -New DataFrame-                      ")
df["Cluster"] = clusters
print(df)
```
