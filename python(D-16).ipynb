#Matplotlib

```python
#creating first chart (Line chart)
print("                                LINE GRAPH                                ")
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y=[100,200,250,300]
plt.xlabel("Years")#for x-axis name
plt.ylabel("Sales")#for y-axis name
plt.title("Sales Report")#Title of Graph
plt.plot(x,y)
plt.show()

print('\n')
print("                                BAR GRAPH                                ")
#creating bar graph
x=[2010,2015,2020,2025]
y=[100,200,250,300]
plt.xlabel("Years")#for x-axis name
plt.ylabel("Sales")#for y-axis name
plt.title("Sales Report")#Title of Graph
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y=[100,200,250,300]
#for graph size
plt.figure(figsize=(6,4))#first is for width and second one is for height
plt.plot(x,y)
plt.show()
```

```python
#customising charts(Line Chart)
from matplotlib import lines
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y=[100,200,250,300]
plt.plot(x,y,color = "red" , marker = "o" , linestyle = "dotted" , linewidth = 4 , markersize = 14)
plt.show()
```

#Multiple Line Chart

```python
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y1=[100,200,260,290]
y2=[150,185,195,300]
plt.plot(x,y1,label = "Jeans")
plt.plot(x,y2,label = "Shirts")
plt.legend() #shows the info about the line in graph
plt.show()
```

#Multiple Bar Chart

```python
from matplotlib import lines
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,30,40]
y2=[20,30,25,30]
w = 0.40
plt.bar(x-w/2,y1,label = "Boys",width = w)
plt.bar(x+w/2,y2,label = "Girls" , width = w)
plt.legend()
plt.xlabel("Group")
plt.ylabel("No. of Students")
plt.title("Number of Students in each Group")
plt.show()
```
