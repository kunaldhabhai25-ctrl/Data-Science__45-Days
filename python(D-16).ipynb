#Matplotlib

```python
#creating first chart (Line chart)
print("                                LINE GRAPH                                ")
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y=[100,200,250,300]
plt.xlabel("Years")#for x-axis name
plt.ylabel("Sales")#for y-axis name
plt.title("Sales Report")#Title of Graph
plt.plot(x,y)
plt.show()

print('\n')
print("                                BAR GRAPH                                ")
#creating bar graph
x=[2010,2015,2020,2025]
y=[100,200,250,300]
plt.xlabel("Years")#for x-axis name
plt.ylabel("Sales")#for y-axis name
plt.title("Sales Report")#Title of Graph
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y=[100,200,250,300]
#for graph size
plt.figure(figsize=(6,4))#first is for width and second one is for height
plt.plot(x,y)
plt.show()
```
