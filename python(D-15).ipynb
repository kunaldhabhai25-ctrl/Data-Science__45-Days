#Removal Of Null Values

Aggregation

```python
import pandas as pd
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file2.json"
df= pd.read_json(url)
print("----------Main Data----------")
print(df)
print("_____________________________________")
print("                  MANUALLY                  ")
print('\n')
print("----------Mean----------")
print(df["marks"].mean())
print('\n')
print("----------Sum----------")
print(df["marks"].sum())
print('\n')
print("----------Min----------")
print(df["marks"].min())
print('\n')
print("----------Max----------")
print(df["marks"].max())
print('\n')
print("----------Count----------")
print(df["marks"].count())

print("_____________________________________")
print("                  USING AGGREGATE FUNCTION                  ")
print('\n')
print(df["marks"].agg(["sum","max","min","count","mean"]))

```

Concat

```python
import pandas as pd
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file2.json"
df= pd.read_json(url)
print("----------Main Data----------")
print(df)
df1 = pd.read_json(url)
print("____________________________________")
print('\n')
#concatanation(by default --> row)
print("----------Cancatanation By row----------")
print(pd.concat([df,df1]))
#concatanation (by column)
print('\n')
print("----------Cancatanation By Column----------")
df2=pd.concat([df,df1],axis=1)
print(df2)
#change the name of abhishek
print('\n')
print("----------Changing Name----------")
df2.loc[2,"name"] = "ankita"
print(df2)
```
