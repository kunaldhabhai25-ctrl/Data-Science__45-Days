#Removal Of Null Values

```python
import pandas as pd
url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/Data-Science__45-Days/main/file2.json"
df= pd.read_json(url)
print("----------Main Data----------")
print(df)
#removal of null values (by default row)
print('\n')
print("----------Removal of null values----------")
print(df.dropna())
#removal of column with null values
print('\n')
print("----------Removal of column with null values----------")
print(df.dropna(axis=1))
```
