```python
#Importing and Using Module's In Python
#example no. --> 1
print("----------Example No. --> 1----------")
print("----------importing module----------")
import math

print(math.cos(44))
print("_____")
print(math.pow(5,3))
print("_____")
print(math.ceil(55.5))
print("_____")
print(math.floor(34.55))
print("_____")
print(math.factorial(7))




#example no. --> 2
print("----------Example No. --> 2----------")
print("----------importing module as name----------")
import math as kunal
print(kunal.degrees(22))

#example no. --> 3
print("----------Example No. --> 3----------")
print("----------importing only pi from module----------")
from math import pi
print(pi*33)
```
