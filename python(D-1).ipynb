CONDITIONAL STATEMENTS IN PYTHON

1. IF
2. ELSE


```python
#if
num1 = 100
num2 = 200
if num1 < num2 :
  print(" num2 is greater : " , num2)
#else
else :
  print("num1 is greater : " , num1)

#else if
name = input("enter your name : ")
age = int(input("enter your age : "))

if age < 18 :
  print("not eligible for voting ")
elif age >= 18 :
  print("name : ", name)
  print("age : ", age)
  print("eligible for voting ")
else :
  print("invalid age ")
```

Example Question

```python
name = input("enter your name : ")
marks = int(input("enter marks : "))
if marks >= 60 and marks <= 74 :
  print("name : ", name)
  print("marks : ", marks)
  print("grade : C ")

elif marks >= 75 and marks <= 89 :
  print("name : ", name)
  print("marks : ", marks)
  print("grade : B ")

elif marks >= 90 :
  print("name : ", name)
  print("marks : ", marks)
  print("grade : A ")

else :
  print("name : ", name)
  print("marks : ", marks)
  print("grade : fail ")
```

```python
L = [10 , 20 , 30 , 44.5 , 77]
for i in range(0,5) :
  print("i : " , i  )
  print("element of list is : " , L[i])


print ('\n\n',"Length of list is : " , len(L))
```
