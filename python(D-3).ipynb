DICTIONARY AND METHODS


```python
dict = {"name":"kunal","Age":"22"}
print(dict.keys())
print(dict.values())
print(dict.items())
print("-----using get-----")
print(dict.get("name"))
print("-----using pop-----")
print(dict.pop("name"))
print(dict)
print("-----using fromkeys-----")
print(dict.fromkeys("Age","22"))
print(dict)
print("-----using update-----")
print(dict.update({"name":"kunal"}))
print(dict)
print("-----using update-----")
print(dict.update({"name":"chocoooo"}))
print(dict)
```

List

```python
#Nested list
#example no. --> 1
print("----------Nested list----------")
L = [10,20,30,["hello","hello1","hello2"],[True,False]]
for i in range(len(L)):
    if type(L[i]) == list:   # check if element is a list
        for j in range(len(L[i])):
            print(L[i][j])


#Slicing in List
print("----------Slicing in List----------")
L = [10,20,30,["hello","hello1","hello2"],[True,False]]
L1=L[1:4]
print(L1)

#Slicing of sublist
print("----------Nested Slicing----------")
L1=L[3:4]
L2=L1[0][1:3]
print(L2)



```

Map , Reduce and Filter


```python
#Function
def sqr(x):
  return x*x

#Map
print("----------Map----------")
L = [1,2,3,4,5]
print(list(map(sqr,L)))

#Lambda
print("----------Lambda----------")
L = [1,2,3,4,5]
print(list(map(lambda x:x*x,L)))

#Reduce
print("----------Reduce----------")
from functools import reduce
L = [1,2,3,4,5]
print(reduce(lambda x,y:x+y,L))


#Filter
print("----------Filter----------")
L = ["apple","banana","cat","dog"]
print(list(filter(lambda x:x[0]=="a",L)))
print(list(filter(lambda x:x.endswith("a"),L)))








```

```python
#example --> try and except
try:
  num1 = 22
  num2 = 3
  print("the division of" , num1 , "by" , num2 , "is : " ,num1/num2)
except:
  print("Error")

#try and except (with error) and finally
try:
  num1 = 10
  num2 = 0

  print(num1/num2)
except:
  print("Error")

finally:
    print("division complete")
```
